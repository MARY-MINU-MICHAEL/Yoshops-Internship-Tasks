### 1.1 Creating Banner

```python
from PIL import Image, ImageDraw, ImageFont, ImageColor

# Using ImageColor to get the color string 
img = ImageColor.getrgb("skyblue")
print(img)
```

```python
# Image.new creates a new image with the given mode and size and this will return an image object.
im = Image.new(mode="RGB", size=(800,800))

im_differ = Image.new(mode = "RGB", size = (900, 150),
                           color = (135, 206, 235))

# Image.open will open the images we want to use in creating the banner
#im1 = Image.open(r"logo.jpg")
im2 = Image.open(r"/content/data-science-python-course.jpg")
im3 = Image.open(r"teen-girl-studying-in-front-computer-flat-vector-33814553.jpg")
im4 = Image.open(r"home-tech-bets-scroll-banner-ai.jpg")
im5 = Image.open(r"QR.png")

# Image.resize(size, resample=0), where the requested size in pixels, as a 2-tuple: (width, height) and PIL.Image.LANCZOS (a high-quality downsampling filter) is an optional resampling filter.
#im1_resized = im1.resize((100, 100), Image.Resampling.LANCZOS)
im2_resized = im2.resize((600, 400), Image.Resampling.LANCZOS)
im3_resized = im3.resize((500, 400), Image.Resampling.LANCZOS)
im5_resized = im5.resize((50, 50), Image.Resampling.LANCZOS)
```

```python
# Image.Image.paste(image1, image2, box), will paste the images where box defines the location of image.
Image.Image.paste(im, im4, (0, 300))
Image.Image.paste(im, im_differ, (0, 400))
Image.Image.paste(im, im3_resized, (-100, 1))
Image.Image.paste(im, im2_resized, (300, 0))
#Image.Image.paste(im, im1_resized, (800, 0))
Image.Image.paste(im, im5_resized, (780, 780))
```

```python
# Defining the font style and size
font = ImageFont.truetype(r'arial.ttf', 40) 
font2 = ImageFont.truetype(r'arial.ttf', 20)
font3 = ImageFont.truetype(r'arial.ttf', 30)
font4 = ImageFont.truetype(r'arial.ttf', 15)

# Defining texts
text = '\n\nBasic Training Program for Everyone!'
text2 = '₹999.00\nOn account of special offer you can get it at ₹899/-\nUse Coupon Code= intern899'
text3 = 'What are you waiting for!'
text4 = 'Go to the link below to book your seats!'
text5 = 'https://yoshops.com/products/data-science-basic-training-program-for-everyone-age-10-to-60'


# Drawing texts
draw = ImageDraw.Draw(im)
draw.text((100, 400), text, fill ="black", font = font, align ="center") 
draw.text((20, 580), text2, fill ="white", font = font2, align ="left") 
draw.text((20, 700), text3, fill ="green", font = font3, align ="left")
draw.text((20, 750), text4, fill ="white", font = font2, align ="left")
draw.text((20, 820), text5, fill ="white", font = font4, align ="left")
```

## 1.3 Creating video

Create and save a video file to `output_path` after 
    combining a static image that is located in `image_path` 
    with an audio file in `audio_path

```python
from moviepy.editor import AudioFileClip, ImageClip

def shortvideomp4(image_path, audio_path, output_path):
    # create the audio clip object
    audio_clip = AudioFileClip('soft_ringtone.mp3')
    
    # create the image clip object
    image_clip = ImageClip('Banner.png')
    
    # use set_audio method from image clip to combine the audio with the image
    video_clip = image_clip.set_audio(audio_clip)
    
    # specify the duration of the new clip to be the duration of the audio clip
    video_clip.duration = audio_clip.duration
    
    # set the FPS to 1
    video_clip.fps = 1
    
    # write the resuling video clip
    video_clip.write_videofile('new.mp4')
    
shortvideomp4('Banner.png', 'soft_ringtone.mp3', 'new.mp4')
```
